```python
import scipy.io as sio
import numpy as np

# Load data
cube = sio.loadmat("Indian_pines_corrected.mat")['indian_pines_corrected']
gt = sio.loadmat("Indian_pines_gt.mat")['indian_pines_gt']

print("Cube shape:", cube.shape)
print("Ground truth shape:", gt.shape)
print("Unique classes (excluding 0):", np.unique(gt)[1:])
```

```python
from skimage.morphology import opening, closing, disk
from tqdm import tqdm
import numpy as np

def morphological_band_scores(cube):
    scores = []
    for b in tqdm(range(cube.shape[2]), desc="Scoring bands"):
        band = cube[:, :, b]
        # Normalize
        band = (band - band.min()) / (band.max() - band.min() + 1e-6)
        opened = opening(band, disk(3))
        closed = closing(band, disk(3))
        morph_response = np.abs(opened - band).mean() + np.abs(closed - band).mean()
        scores.append((b, morph_response))
    # Sort descending by score
    sorted_bands = sorted(scores, key=lambda x: x[1], reverse=True)
    return [b for b, _ in sorted_bands[:15]]  # top 15 bands

top15 = morphological_band_scores(cube)
print("Top 15 band indices:", top15)
```

```python
from sklearn.model_selection import train_test_split

def extract_patches(cube, gt, band_indices, patch_size=7):
    margin = patch_size // 2
    H, W, _ = cube.shape

    # Extract the selected bands
    cube = cube[:, :, band_indices]

    X, y = [], []
    for i in range(margin, H - margin):
        for j in range(margin, W - margin):
            label = gt[i, j]
            if label == 0:
                continue
            patch = cube[i - margin:i + margin + 1, j - margin:j + margin + 1, :]
            X.append(patch)
            y.append(label)
    return np.array(X), np.array(y)

X, y = extract_patches(cube, gt, top15, patch_size=7)
print("X shape:", X.shape)
print("y shape:", y.shape)

# Normalize input
X = X.astype(np.float32)
X = (X - X.min()) / (X.max() - X.min())

# Stratified 30% train split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.7, random_state=0, stratify=y
)

print("Train samples:", X_train.shape[0])
print("Test samples:", X_test.shape[0])
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import LabelEncoder

# Encode labels from 1–16 ➝ 0–15
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)

class HsiDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.from_numpy(X).permute(0, 3, 1, 2)  # (N, C, H, W)
        self.y = torch.from_numpy(y).long()

    def __len__(self):
        return self.X.shape[0]

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

train_ds = HsiDataset(X_train, y_train_enc)
test_ds = HsiDataset(X_test, y_test_enc)

train_loader = DataLoader(train_ds, batch_size=64, shuffle=True)
test_loader = DataLoader(test_ds, batch_size=64, shuffle=False)
```

```python
import torch.nn as nn
import torch.nn.functional as F

class Baseline3DCNN(nn.Module):
    def __init__(self, input_channels=15, n_classes=16):
        super(Baseline3DCNN, self).__init__()
        self.conv1 = nn.Conv3d(1, 8, kernel_size=(3, 3, 3), padding=1)
        self.bn1 = nn.BatchNorm3d(8)
        self.conv2 = nn.Conv3d(8, 16, kernel_size=(3, 3, 3), padding=1)
        self.bn2 = nn.BatchNorm3d(16)
        self.fc1 = nn.Linear(16 * 7 * 7 * input_channels, 128)
        self.dropout = nn.Dropout(0.5)
        self.fc2 = nn.Linear(128, n_classes)

    def forward(self, x, return_features=False):
        x = x.unsqueeze(1)
        x = F.relu(self.bn1(self.conv1(x)))
        x = F.relu(self.bn2(self.conv2(x)))
        x = x.view(x.size(0), -1)
        x = F.relu(self.fc1(x))
        if return_features:
          return x  # return penultimate layer activations
        x = self.dropout(x)
        return self.fc2(x)
```

```python
import torch.optim as optim

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = Baseline3DCNN().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

def train(model, loader, optimizer, criterion):
    model.train()
    total_loss = 0
    for X_batch, y_batch in loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        optimizer.zero_grad()
        out = model(X_batch)
        loss = criterion(out, y_batch)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(loader)

def evaluate(model, loader):
    model.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for X_batch, y_batch in loader:
            X_batch = X_batch.to(device)
            out = model(X_batch)
            preds = torch.argmax(out, dim=1).cpu().numpy()
            y_true.extend(y_batch.numpy())
            y_pred.extend(preds)
    return y_true, y_pred

# Train
n_epochs = 30
for epoch in range(n_epochs):
    loss = train(model, train_loader, optimizer, criterion)
    print(f"Epoch {epoch+1}/{n_epochs}, Loss: {loss:.4f}")

# Evaluate
y_true, y_pred = evaluate(model, test_loader)
print("\n✅ Classification Report (Baseline):")
print(classification_report(y_true, y_pred, digits=4))
```

```python
import torch.nn as nn
import torch.nn.functional as F

class Hybrid3DCNN(nn.Module):
    def __init__(self, input_channels=15, n_classes=16):
        super(Hybrid3DCNN, self).__init__()
        self.conv1 = nn.Conv3d(1, 8, kernel_size=(3, 3, 3), padding=1)
        self.bn1 = nn.BatchNorm3d(8)
        self.conv2 = nn.Conv3d(8, 16, kernel_size=(3, 3, 3), padding=1)
        self.bn2 = nn.BatchNorm3d(16)

        # Your patch size is (7x7x15), after 3D conv: (B, 16, 7, 7, 15)
        # For attention, we flatten spatial (C×H×W) = 16×7×7 = 784 per band
        self.embed_dim = 16 * 7 * 7  # 784
        # Set batch_first=False for MultiheadAttention
        self.attn = nn.MultiheadAttention(embed_dim=self.embed_dim, num_heads=4, batch_first=False)

        self.fc1 = nn.Linear(input_channels * self.embed_dim, 128)  # 15 × 784 = 11760
        self.dropout = nn.Dropout(0.5)
        self.fc2 = nn.Linear(128, n_classes)

    def forward(self, x, return_features=False):
        x = x.unsqueeze(1)  # (B, 1, 15, 7, 7) ➝ (B, 1, C, H, W)
        x = F.relu(self.bn1(self.conv1(x)))   # ➝ (B, 8, 15, 7, 7)
        x = F.relu(self.bn2(self.conv2(x)))   # ➝ (B, 16, 15, 7, 7)

        B, C_out, D, H, W = x.shape  # (B, 16, 15, 7, 7)
        x = x.permute(2, 0, 1, 3, 4)          # ➝ (D, B, C_out, H, W)
        x = x.reshape(D, B, -1)               # ➝ (D, B, C_out*H*W) = (15, B, 784)

        # Attention expects (seq_len, B, embed_dim) where seq_len = D = 15 and embed_dim = C_out*H*W = 784
        attn_out, _ = self.attn(x, x, x)      # ➝ (15, B, 784)
        attn_out = attn_out.permute(1, 0, 2)  # ➝ (B, 15, 784)
        attn_out = attn_out.reshape(B, -1)    # ➝ (B, 15 × 784 = 11760)

        x = F.relu(self.fc1(attn_out))
        if return_features:
            return x
        x = self.dropout(x)
        return self.fc2(x)
```

```python
hybrid_model = Hybrid3DCNN().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(hybrid_model.parameters(), lr=1e-3)
```

```python
n_epochs = 30
for epoch in range(n_epochs):
    loss = train(hybrid_model, train_loader, optimizer, criterion)
    print(f"Hybrid Epoch {epoch+1}/{n_epochs}, Loss: {loss:.4f}")
```

```python
y_true_hybrid, y_pred_hybrid = evaluate(hybrid_model, test_loader)
print("\n✅ Classification Report (Hybrid):")
print(classification_report(y_true_hybrid, y_pred_hybrid, digits=4))
```

```python
from tqdm import tqdm
from sklearn.metrics import classification_report

# Define constants
PATCH_SIZE = 7
PAD_SIZE = PATCH_SIZE // 2
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ---- Step 1: Pad the hyperspectral image ----
def pad_data(data, pad_size):
    return np.pad(data, ((pad_size, pad_size), (pad_size, pad_size), (0, 0)), mode='constant')

# ---- Step 2: Extract a patch ----
def extract_patch(data, i, j, patch_size):
    return data[i:i+patch_size, j:j+patch_size, :].transpose(2, 0, 1)  # (C, H, W)
```

```python
# ---- Step 3: Load the data you used ----
hyperspectral_data = cube[:, :, top15]  # selected 15 bands
ground_truth = gt  # integer labels, shape (145, 145)
```

```python
# ---- Step 4: Prepare padded data ----
padded_data = pad_data(hyperspectral_data, PAD_SIZE)
H, W = ground_truth.shape
predicted_map = np.zeros((H, W), dtype=int)
```

```python
# ---- Step 5: Load trained model ----
hybrid_model = Hybrid3DCNN(input_channels=15, n_classes=16).to(device)
hybrid_model.load_state_dict(torch.load("hybrid_model.pth", map_location=device))
hybrid_model.eval()
```

```python
# ---- Step 6: Predict each pixel ----
with torch.no_grad():
    for i in tqdm(range(H)):
        for j in range(W):
            if ground_truth[i, j] == 0:
                continue  # Skip background

            patch = extract_patch(padded_data, i, j, PATCH_SIZE)
            patch = torch.from_numpy(patch).float().unsqueeze(0).to(device)  # (1, C, H, W)
            #patch = extract_patch(padded_data, i, j, PATCH_SIZE)  # (C, H, W)
            #patch = torch.from_numpy(patch).float().unsqueeze(0).unsqueeze(0).to(device)  # (1, 1, C, H, W)
            output = hybrid_model(patch)
            pred = output.argmax(dim=1).item()
            predicted_map[i, j] = pred + 1  # +1 to match GT indexing
```

```python
# Step 7: Visualize
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(ground_truth, cmap='tab20')
plt.title("Ground Truth")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(predicted_map, cmap='tab20')
plt.title("Hybrid Model Prediction")
plt.axis('off')

plt.tight_layout()
plt.savefig("hybrid_prediction_vs_gt.png")
plt.show()
```

```python
# Step 8: Evaluate accuracy (only on labeled pixels)
mask = ground_truth > 0
y_true = ground_truth[mask] - 1
y_pred = predicted_map[mask] - 1

print("✅ Final Report on Full Image:")
print(classification_report(y_true, y_pred, digits=4))

oa = accuracy_score(y_true, y_pred)  # already filtered and -1 adjusted

print(f"\n✅ Overall Accuracy (Hybrid): {oa*100:.2f}%")
```

```python
from sklearn.metrics import cohen_kappa_score, f1_score

def compute_metrics(y_true, y_pred):
    oa = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred, average=None)
    aa = np.mean(f1_score(y_true, y_pred, average=None))
    kappa = cohen_kappa_score(y_true, y_pred)
    print(f"\n✅ Final Hybrid Model Metrics:")
    print(f"Overall Accuracy (OA): {oa*100:.2f}%")
    print(f"Average Accuracy (AA): {aa*100:.2f}%")
    print(f"Kappa Score: {kappa:.4f}")
    print(f"Per-Class F1: {f1}")
```
